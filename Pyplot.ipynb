```python
# Biblioteka za prikaz grafikona
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Prikaz f-je

x = np.linspace(1, 3, 100)
y0 = x
y1 = x * np.log(x)
y2 = np.sqrt(x)
y4 = x * x
y5 = x * x * x

plt.plot(x,y0)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y4)
plt.plot(x,y5)

plt.xlabel("x")
plt.ylabel("y")
plt.legend([
    "y = x",
    "y = $x * log(x)$",
    "y = $\sqrt{x}$",
    "y = $x ^ 2$",
    "y = $x ^ 3$"
])

plt.savefig("data/proba.pdf")
plt.show()
```
